```python
#@title Calculating rmsd {display-mode: "form"}
import os
from spyrmsd import io, rmsd
from spyrmsd.optional import obabel as ob
from spyrmsd import molecule
import numpy as np

energy_optimizer = 'mma'

#index_array = np.array(df.index)
target_cut = targets[0:10]
for target in target_cut:

  folder = sb_folder+target


  %cd $folder
  subfolder = "self_docking_"+energy_optimizer
  ! mkdir -p $subfolder
  %cd $subfolder


  if "McLiBELa_dock.mol2.gz" in os.listdir():
    !rm -rf McLiBELa.mol2 ligref.mol2

    !cp lig.mol2.gz ligref.mol2.gz
    !gunzip -f ligref.mol2.gz

    obmol_ref = ob.load("ligref.mol2")
    mol_ref = ob.to_molecule(obmol_ref)
    mol_ref.strip()
    print(io.numatoms(obmol_ref))

    !cp McLiBELa_dock.mol2.gz McLiBELa.mol2.gz
    !gunzip -f McLiBELa.mol2.gz

    obmol_dock = ob.load("McLiBELa.mol2")
    mol_dock = ob.to_molecule(obmol_dock)
    mol_dock.strip()
    print(io.numatoms(obmol_dock))



    if io.numatoms(obmol_ref) == io.numatoms(obmol_dock):
      coords_dock = molecule.coords_from_molecule(mol_dock)
      atomicn_dock = mol_dock.atomicnums

      coords_ref = molecule.coords_from_molecule(mol_ref)
      atomicn_ref = mol_ref.atomicnums

      h = rmsd.hrmsd(coords_dock, coords_ref, atomicn_dock, atomicn_ref)
      print(h)
    #print(rmsd.symmrmsd(mol1, mol2))



#por algum motivo tem um ligante diferente nos 5 primeiros
```

```python
#@title Reading the Docks {display-mode: "form"}
import os

energy_optimizer = 'mma'



header = "target\tREC.N\tRefLig.N\tLig2.N\tstart_energy\ttime_grid\tgrid_energy\ttime_dock\n"
print(header)

all_files = header

for j in range(700):
  print(j)
  target = targets[j]

  folder = sb_folder+target

#  print()
#  print()
  %cd $folder
  subfolder = "self_docking_"+energy_optimizer
  ! mkdir -p $subfolder
  %cd $subfolder


  if "log.txt" in os.listdir():

      log = open('log.txt')
      data = log.readlines()
      if len(data) == 2:
        data = data[1]
        print(data)

        all_files+=data


```

```python
#@title Generating datasheet {display-mode: "form"}

import numpy as np
import pandas as pd

all_files_lines = all_files.split('\n')


print(all_files_lines[-1])
N=len(all_files.split('\n'))
datasheet = np.zeros((N,7))

k=0

target_list = N*['']

for line in all_files_lines[1:]:
  if len(line) !=0:
    listvar = [float(i) for i in line.split()[1:]]
    #print(listvar)
    datasheet[k] = listvar

    target_list[k] = line[:4]
    k+=1


header = header.strip("\n")
header_list = header.split("\t")[1:]
df = pd.DataFrame(datasheet, index = target_list, columns=header_list)
df
```
