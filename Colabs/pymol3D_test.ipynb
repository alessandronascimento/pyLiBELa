```python
#@title Arrumando arquivos pdb {display-mode: "form"}

%%capture
from google.colab import drive
drive.mount('/content/drive/')
path = '/content/drive/MyDrive/pyLiBELa/SB/'

pdb_code = "182L" #@param {type:"string"}

#copiando pro colab os arquivos do drive do código pdb dado
folder = path+pdb_code+'/*'
!mkdir $pdb_code
!cp $folder $pdb_code/
!ls $pdb_code

#convertendo de mol2 pra pdb
name_file_rec = pdb_code+'_rec.pdb'
name_src_rec = pdb_code + '/' + pdb_code + '.rec.clean.mol2.gz'
!obabel -imol2 $name_src_rec -opdb -O $name_file_rec

name_file_lig = pdb_code+'_lig.pdb'
name_src_lig = pdb_code + '/' + pdb_code + '.lig.am1bcc.mol2.gz'
!obabel -imol2 $name_src_lig -opdb -O $name_file_lig

!sed -i '/END\|CONECT/d' $name_file_rec #tira o end e o conect do arquivo do receptor

!sed -i 's/A  /B  /g' $name_file_lig #troca a cadeia A por B no arquivo do ligante

!sed -i '/CONECT/d' $name_file_lig #tira o conect do arquivo do ligante

name_file_merge = pdb_code + '.pdb'
!cat $name_file_rec $name_file_lig > $name_file_merge #concatena os arquivos
```

```python
#Visualização
view = py3Dmol.view()

#Adicionando os modelos
view.addModel(open(name_file_merge, 'r').read(),'pdb') 


view.addModel(open(pdb_code+'/box.pdb', 'r').read(),'pdb') 


#Ajustando a visualização
view.setStyle({'chain':"B"},{'stick': {'colorscheme':'yellowCarbon'}})


inorganic = ['Mg','Ca'] #compostos inorganicos, n são tantos assim, né?

#view.setStyle({'model':1}, {'stick':{"color":"#ff0000","width":1}}) #visualização do box mais espesso

view.setStyle({'chain':'A'},  {"cartoon":{'arrows':True,'colorscheme':'ssPymol','opacity':0.7}})
view.setStyle({'elem':inorganic}, {'sphere':{}}) 

view.addStyle({'chain':'A','within':{'distance':'5', 'sel':{'chain':'B'}}}, {'stick': {}}) 



view.setBackgroundColor('white')
view.zoomTo({"model":1})
view.show()
```
